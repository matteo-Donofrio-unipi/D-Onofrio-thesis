```python
from binarytree import Node
import pandas as pd
import numpy as np
from sklearn import tree
from scipy.io import arff
```

```python
def printAll(Root):
    print('Nodo: '+str(Root.value))
    print(Root.data)
    if(Root.left!=None):
        printAll(Root.left)
    if(Root.right!=None):
        printAll(Root.right)
```

```python
dataset = arff.loadarff('CBF/CBF_TRAIN.arff')
df = pd.DataFrame(dataset[0]) #30 record su matrice da 128 attributi + 'b': classe appartenenza  
df.shape
#print(df)
Ts0 = np.array(df.iloc[0].values)
Ts1 = np.array(df.iloc[1].values)
Ts2 = np.array(df.iloc[2].values)
```

```python
root = Node(0)
root.data=df
root.left = Node(1)
root.left.data=df.iloc[0:2]
root.right = Node(2)
root.right.data=df.iloc[2]
print(root)
printAll(root)
```
